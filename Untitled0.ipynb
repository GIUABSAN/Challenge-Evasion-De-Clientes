```python
import pandas as pd
import requests

# URL del archivo JSON raw en GitHub
json_url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Descargar el contenido del JSON
response = requests.get(json_url)

# Verificar si la descarga fue exitosa
if response.status_code == 200:
    data = response.json()
    # Convertir los datos a un DataFrame de Pandas
    df_churn = pd.DataFrame(data)

    print("Datos cargados exitosamente desde la API.")
    # Mostrar las primeras filas del DataFrame
    display(df_churn.head())

    # Mostrar información del DataFrame (columnas, tipos de datos, valores no nulos)
    display(df_churn.info())
else:
    print(f"Error al descargar los datos. Código de estado: {response.status_code}")
    df_churn = None
```

```python
# Flatten the 'customer' column
customer_df = pd.json_normalize(df_churn['customer'])
df_churn = pd.concat([df_churn.drop('customer', axis=1), customer_df], axis=1)

# Flatten the 'phone' column
phone_df = pd.json_normalize(df_churn['phone'])
df_churn = pd.concat([df_churn.drop('phone', axis=1), phone_df], axis=1)

# Flatten the 'internet' column
internet_df = pd.json_normalize(df_churn['internet'])
df_churn = pd.concat([df_churn.drop('internet', axis=1), internet_df], axis=1)

# Flatten the 'account' column
account_df = pd.json_normalize(df_churn['account'])
df_churn = pd.concat([df_churn.drop('account', axis=1), account_df], axis=1)

print("Columnas anidadas expandidas.")
# Display the first few rows of the updated DataFrame
display(df_churn.head())

# Display information of the updated DataFrame
display(df_churn.info())
```

```python
# Check for missing values in the flattened DataFrame
print("Missing values per column:")
display(df_churn.isnull().sum())

# Display data types of the columns in the flattened DataFrame
print("\nData types of columns:")
display(df_churn.dtypes)
```

```python
# Attempt to convert 'Charges.Total' to numeric, coercing errors to NaN
df_churn['Charges.Total'] = pd.to_numeric(df_churn['Charges.Total'], errors='coerce')

# Check for NaN values introduced by the coercion
print("Number of NaN values in 'Charges.Total' after coercion:")
display(df_churn['Charges.Total'].isnull().sum())

# Display the data type of 'Charges.Total' after conversion
print("\nData type of 'Charges.Total' after conversion:")
display(df_churn['Charges.Total'].dtype)
```

## Conclusiones e Insights

Basado en el análisis exploratorio de datos, podemos extraer los siguientes insights clave sobre la evasión de clientes en este dataset:

*   **La antigüedad es un predictor fuerte:** Los clientes nuevos o con poca antigüedad son los más vulnerables a la evasión.
*   **El tipo de contrato es crucial:** Los contratos mes a mes están fuertemente asociados con una alta tasa de evasión, mientras que los contratos a largo plazo fomentan la retención.
*   **Servicios adicionales y soporte técnico importan:** Los clientes que no tienen servicios de seguridad, respaldo, protección o soporte técnico en línea son más propensos a irse. Esto sugiere que estos servicios pueden aumentar la percepción de valor y la lealtad.
*   **El servicio de Internet de Fibra Óptica y el Pago Electrónico:** Estos parecen ser puntos de fricción o indicadores de clientes con mayor propensión a evadir. Podría haber problemas de calidad con el servicio de fibra óptica o la facilidad de cancelar para quienes pagan electrónicamente.
*   **Ciudadanos Mayores y Clientes sin Dependientes:** Estos segmentos muestran una mayor tasa de evasión, lo que podría indicar necesidades o preferencias diferentes que no están siendo completamente satisfechas.
*   **Cargos Mensuales:** Aunque los cargos totales son menores para los clientes que evaden (debido a la antigüedad), sus cargos mensuales o diarios son ligeramente más altos, posiblemente relacionado con la contratación de servicios más costosos (como fibra óptica) que quizás no cumplen sus expectativas.

Estos hallazgos sugieren que la evasión no se debe a un único factor, sino a una combinación de antigüedad, tipo de contrato, servicios contratados y métodos de pago.
